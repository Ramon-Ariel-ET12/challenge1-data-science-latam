### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Agregar una columna para identificar la tienda
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Unir todos los DataFrames
datos = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)


datos.head()
```

# 1. Análisis de facturación



```python
 # Asumimos que 'Precio' es el valor total de cada compra
datos['Total'] = datos['Precio']
facturacion = datos.groupby('Tienda')['Total'].sum().sort_values(ascending=False)
print(facturacion)
```

# 2. Ventas por categoría

```python
# Agrupar por tienda y categoría, sumando las ventas (Cantidad de cuotas)
categorias_populares = datos.groupby(['Tienda', 'Categoría del Producto'])['Cantidad de cuotas'].sum().reset_index()

# Ordenar por tienda y cantidad de ventas, de mayor a menor
categorias_top = categorias_populares.sort_values(['Tienda', 'Cantidad de cuotas'], ascending=[True, False])

# Mostrar las categorías más populares por tienda
categorias_top.groupby('Tienda').head(1)
```

# 3. Calificación promedio de la tienda


```python
evaluacion_promedio = datos.groupby('Tienda')['Calificación'].mean().sort_values(ascending=False)
print(evaluacion_promedio)
```

# 4. Productos más y menos vendidos

```python
productos_vendidos = datos.groupby(['Tienda', 'Producto']).size().reset_index(name='Ventas')
productos_ordenados = productos_vendidos.sort_values(['Tienda', 'Ventas'], ascending=[True, False])

# Producto más vendido
productos_mas_vendidos = productos_ordenados.groupby('Tienda').head(1)

# Producto menos vendido
productos_menos_vendidos = productos_ordenados.groupby('Tienda').tail(1)

print("Más vendidos:")
print(productos_mas_vendidos)
print("\nMenos vendidos:")
print(productos_menos_vendidos)
```

# 5. Envío promedio por tienda

```python
costo_envio_promedio = datos.groupby('Tienda')['Costo de envío'].mean().sort_values(ascending=False)
print(costo_envio_promedio)
```
